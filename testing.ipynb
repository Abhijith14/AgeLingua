```python
dialogue_data2 = []
for path in f_paths2:  # for each of the dialogues
    for i in get_xml(path).xpath('//u'):
        u_n = i.xpath('./@n')[0]
        u_who = i.xpath('./@who')[0]
        u_trans = i.xpath('./@trans')[0]
        u_confidence = i.xpath('./@whoConfidence')[0]
    
        word_data = []
        for w in i.xpath('.//w'):
            w_pos = w.xpath('./@pos')[0]
            w_lemma = w.xpath('./@lemma')[0]
            w_class = w.xpath('./@class')[0]
            w_usas = w.xpath('./@usas')[0]
            w_words = w.xpath('./text()')[0]

            word_data.append({
                'u_n': u_n,
                'u_who': u_who,
                'u_trans': u_trans,
                'u_confidence': u_confidence,
                'w_pos': w_pos,
                'w_lemma': w_lemma,
                'w_class': w_class,
                'w_usas': w_usas,
                'w_words': w_words,
                'dialogue': path.split('/')[-1].replace("-tgd.xml", "")
            })

        dialogue_data2.extend(word_data)

df_word_data = pd.DataFrame(dialogue_data2)
```

```python
# untagged

def get_xml(f_path):
    with open(f_path, 'r') as f:
        f = f.read()
    xml = etree.fromstring(f)
    return xml
```

```python
# delete cell

# f_paths = ['bnc2014spoken-xml/spoken/untagged/S2A5.xml']

f_paths
```

```python
dialogue_data = []
for path in f_paths:  # for each of the dialogues
    for i in get_xml(path).xpath('//u'):
        u = i.xpath('./text()')
        n = i.xpath('./@n')[0]
        who = i.xpath('./@who')[0]
        trans = i.xpath('./@trans')
        # check if i contains the tag pause. If it does, then get the value of the pause attribute
        if i.xpath('./pause'):
            pause = i.xpath('./pause/@dur')
            pause = ', '.join(pause)
        else:
            pause = None

        # check if i contains the tag vocal. If it does, then get the value of the vocal attribute
        if i.xpath('./vocal'):
            vocal = i.xpath('./vocal/@desc')
            vocal = ', '.join(vocal)
        else:
            vocal = None

        # check if i contains the tag anon. If it does, then get the value of the anon attribute
        if i.xpath('./anon'):
            anonType = i.xpath('./anon/@type')
            anon_nameType = i.xpath('./anon/@nameType')
            anonType = ', '.join(anonType)
            anon_nameType = ', '.join(anon_nameType)
        else:
            anonType = None
            anon_nameType = None
        
        # check if i contains the tag unclear. If it does, then get the value of the unclear attribute
        if i.xpath('./unclear'):
            unclear = True
        else:
            unclear = False

        # check if i contains the tag trunc. If it does, then get the value of the trunc attribute
        if i.xpath('./trunc'):
            trunc = i.xpath('./trunc/./text()')[0]
        else:
            trunc = None

        # check if i contains the tag event. If it does, then get the value of the event attribute
        if i.xpath('./event'):
            event = i.xpath('./event/@desc')
            event = ', '.join(event)
        else:
            event = None

        # convert all lists to strings with coma separated values
        u = ', '.join(u)
        trans = ', '.join(trans)

        if trans == '':
            trans = None

        dialogue_data.append({
                'u': u,
                'n': n,
                'who': who,
                'trans': trans,
                'pause': pause,
                'vocal': vocal,
                'anonType': anonType,
                'anon_nameType': anon_nameType,
                'unclear': unclear,
                'trunc': trunc,
                'event': event,
                'dialogue': path.split('/')[-1].replace(".xml", "")
        })
```

```python
# convert the list of dictionaries into a dataframe
df_dialogue_data = pd.DataFrame(dialogue_data)
df_dialogue_data
```
